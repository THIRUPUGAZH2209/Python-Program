# Arithmetic Operators

```python
a=int(input("Enter first number"))
b=int(input("Enter second number"))
print("Addition:",a+b)
print("Subtraction:",a-b)
print("Multiplication:",a*b)
print("Divison:",a/b)
print("Modulus:",a%b)
print("Exponent:",a**b)
```

# Lists, Tuples, Sets and Dictionaries

```python
# list
my_list=[1,2,3]
my_list.append(4)

#Tuple
my_tuple=(1,2,3)

#Set
my_set={1,2,3}
my_set.add(4)

# Dictionary
my_dict={'name':"Thiru","Age":23}
my_dict['city']='New York'

print("List:",my_list)
print("Tuple:",my_tuple)
print("Set:",my_set)
print("Dictionary:",my_dict)
```

# Looping Constructs and Iterators

```python
# For Loop
for i in range(5):
    print("For loop:",i)

# While Loop
i=0
while i<5:
    print("while Loop:",i)
    i+=1

# Iterator
a=[1,2,3]
b=iter(a)
print(next(b))
print(next(b))
print(next(b))
```

# Functios and Lambda Functions

```python
# Function
def square(x):
    return x**2
print("Square value is",square(5))

# Lambda Function
cube=lambda x:x**3
print("Cube:",cube(3))
```

# Simple File Handling

```python
# Writing to a file
with open('sample.txt', 'w') as file:
 file.write('Hello World!')
# Reading the file
with open('sample.txt', 'r') as file:
 print(file.read())
# Appending to the file
with open('sample.txt', 'a') as file:
 file.write('\nAppended text.')
# Reading the file
with open('sample.txt', 'r') as file:
 print(file.read())
```

# Object-Oriented Programming(OOP)

```python
class Animal:
 def __init__(self, name):
     self.name = name
 def speak(self):
     print(self.name, "says Hello!")
class Dog(Animal):
 def speak(self):
     print(self.name, "says Woof!")
a = Animal("Animal")
a.speak()
d = Dog("Dog")
d.speak()
```

# Python Libraries(Numpy,Pandas,Matplotlib)
