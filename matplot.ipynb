```python
import pandas as pd
data={'Name':['alice','bob','charlie'],
      'Age':[24,27,22],
      'City':['Newyork','Los Angeles',"Chicago"]}
df=pd.DataFrame(data)
df.to_csv("sample")
print(df)
```

# Reading CSV Files

```python
df=pd.read_csv('sample')
print(df.head())
```

# Data cleaning

```python
df=pd.DataFrame({'A':[1,2,None,4],"b":[None,2,3,4]})
df_cleaned=df.dropna()
print(df_cleaned)
```

# Filtering Data

```python
df=pd.DataFrame({'a':[1,2,3,4],'b':[5,6,7,8]})
filtered_df=df[df['a']>2]
print(filtered_df)
```

# Grouping Data

```python
df=pd.DataFrame({'a':['foo','bar','foo','bar'],'b':[1,2,3,4]})
grouped=df.groupby('a').sum()
print(grouped)
```

# Mergind DataFrames

```python
df1=pd.DataFrame({'a':['foo','bar'],'b':[1,2]})
df2 = pd.DataFrame({'a': ['foo', 'bar'], 'c': [3, 4]})
merged_df = pd.merge(df1, df2, on='a')
print(merged_df)
```

# Pivot Tables

```python
df=pd.DataFrame({'a':['foo','bar','foo','bar'],
                  'b':[1,2,3,4],
                  'c':[5,6,7,8]})
pivot_tables=df.pivot_table(values='c',index='a',columns='b', aggfunc='sum')
print(pivot_tables)
```

# DataFrame Transformation

```python
df = pd.DataFrame({'a': [1, 2, 3, 4], 'b': [5, 6, 7, 8]})
transformed_df = df.transform(lambda x: x ** 2)
print(transformed_df)
```

# Handling Duplicates

```python
df = pd.DataFrame({'a': [1, 1, 2, 3], 'b': [4, 4, 5, 6]})
df_no_duplicates = df.drop_duplicates()
print(df_no_duplicates)
```

 # Saving DataFrames to CSV

```python
df = pd.DataFrame({'a': [1, 2, 3], 'b': [4, 5, 6]})
df.to_csv('output.csv', index=False)
print("DataFrame saved to output.csv")
```
